```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

# - - - - - - - - 座標軸の設定 - - - - - - - - #

# 軸を中心にする
ax.spines["bottom"].set_position("zero")
ax.spines["left"].set_position("zero")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# メモリのスタイル
ax.tick_params(direction="inout")

# メモリの数値設定
ax.set_xticks([-1, -0.5, 0, 0.5, 1])
ax.set_xticklabels([-1, -0.5, 0, 0.5, 1])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])
ax.set_yticklabels([-1, -0.5, "", 0.5, 1])

# グラフの範囲
ax.set_xlim([-1.1, 1.1])
ax.set_ylim([-1.1, 1.1])

# 軸ラベル
ax.set_xlabel("第1因子負荷量", rotation=-90, family="osaka", x=0.98)
ax.set_ylabel("第2因子負荷量", rotation=0, family="osaka", y=0.97)


# - - - - - - - - 円の描画 - - - - - - - - #

c = patches.Circle(xy=(0, 0), radius=1, fill=False)
ax.add_patch(c)

# - - - - - - - - プロットの設定 - - - - - - - - #

# 点のラベル
data_name = ["数学", "国語", "理科"]

# プロットするデータ
f1 = [0.91, 0.88, 0.83]
f2 = [-0.15, -0.36, 0.54]

# プロット
for i, j, k in zip(f1, f2, data_name):
    plt.plot(i, j, "o")
    plt.annotate(k, xy=(i, j), family="osaka")

plt.show()
```
